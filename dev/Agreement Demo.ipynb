```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import ascii
```

```python
data1 = ascii.read("sample_output_heal.txt")  
plt.plot(data1['col1'], data1['col1']**2 * data1['col2'], label="healpix")

data2 = ascii.read("sample_output.txt")
plt.plot(data2['col1'], data2['col1']**2 * data2['col2'], lw=1.5, label="CAR", color="red")

plt.ylabel(r"$\ell^2 C_{\ell}$")
plt.ylim(-0.005, 0.04)
plt.xlim(0,2000)
# plt.yscale("log")
plt.legend()
plt.xlabel(r"$\ell$")
```
